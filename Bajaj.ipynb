```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
import json
data = json.load(open('/content/DataEngineeringQ2.json'))
```

```python
# Converting JSON data to a DataFrame
df = pd.json_normalize(data)
df
```

```python
# Extract patient details
patient_details = [record["patientDetails"] for record in data]

# Initialize counters
total_records = len(patient_details)
missing_first_name = 0
missing_last_name = 0
missing_dob = 0

# Count missing values for each field
for details in patient_details:
    if "firstName" not in details or not details["firstName"]:
        missing_first_name += 1
    if "lastName" not in details or not details["lastName"]:
        missing_last_name += 1
    if "birthDate" not in details or not details["birthDate"]:
        missing_dob += 1

# Calculate percentages
missing_first_name_percent = (missing_first_name / total_records) * 100
missing_last_name_percent = (missing_last_name / total_records) * 100
missing_dob_percent = (missing_dob / total_records) * 100

# Round off to 2 decimal places
missing_first_name_percent = round(missing_first_name_percent, 2)
missing_last_name_percent = round(missing_last_name_percent, 2)
missing_dob_percent = round(missing_dob_percent, 2)

# Format the result
result = f"{missing_first_name_percent}, {missing_last_name_percent}, {missing_dob_percent}"
result
```

```python

# Imputing gender column based on mode
mode_gender = df['patientDetails.gender'].mode()[0]
df['patientDetails.gender'].fillna(mode_gender, inplace=True)

# Calculating the percentage of female gender
total_gender_entries = df['patientDetails.gender'].count()
female_percentage = (df['patientDetails.gender'].value_counts(normalize=True)['F']) * 100
female_percentage = round(female_percentage, 2)

print(female_percentage)
```

```python

from datetime import datetime

# Calculating Age
df['patientDetails.birthDate'] = pd.to_datetime(df['patientDetails.birthDate'], errors='coerce')
df['age'] = df['patientDetails.birthDate'].apply(lambda x: datetime.now().year - x.year if pd.notnull(x) else None)

# Adding Age Group column
df['ageGroup'] = pd.cut(df['age'], bins=[0, 12, 19, 59, 120], labels=['Child', 'Teen', 'Adult', 'Senior'])

# Counting the number of Adults
adult_count = df['ageGroup'].value_counts().get('Adult', 0)
print(adult_count)
```

```python

# Loading consultationData.medicines into a separate DataFrame
med_df = pd.json_normalize(data, record_path=['consultationData', 'medicines'], meta=['appointmentId'])

# grouping appointmentId and counting medicines prescribed
medicine_count_per_appointment = med_df.groupby('appointmentId').size()

# Calculating average number of medicines prescribed
average_medicines = medicine_count_per_appointment.mean()
average_medicines = round(average_medicines, 2)

print(average_medicines)
```

```python
# Finding the 3rd most frequently prescribed medicineName
medicine_counts = med_df['medicineName'].value_counts()
third_most_prescribed_medicine = medicine_counts.index[2]  # for the 3rd item index will be 2 starting from 0

print(third_most_prescribed_medicine)
```

```python
# Calculating the distribution of active and inactive medicines
total_medicines = med_df['isActive'].count()
active_medicines = med_df['isActive'].sum()
inactive_medicines = total_medicines - active_medicines

active_percentage = (active_medicines / total_medicines) * 100
inactive_percentage = (inactive_medicines / total_medicines) * 100

active_percentage = round(active_percentage, 2)
inactive_percentage = round(inactive_percentage, 2)

print(f"Active: {active_percentage}%, Inactive: {inactive_percentage}%")
```

```python
import regex as re
# Defining the validation function
def is_valid_indian_phone_number(phone):
    # Removing any spaces or non-numeric characters
    phone = re.sub(r'\D', '', phone)

    # Checking for valid prefixes and remove them
    if phone.startswith('91'):
        phone = phone[2:]
    elif phone.startswith('+91'):
        phone = phone[3:]

    # Checking if the remaining part is a valid 10-digit number
    if len(phone) == 10 and phone.isdigit():
        phone_num = int(phone)
        if 6000000000 <= phone_num <= 9999999999:
            return True
    return False

# Applying the validation function to the phoneNumber column
df['isValidMobile'] = df['phoneNumber'].apply(is_valid_indian_phone_number)

# Counting the number of valid phone numbers
valid_phone_numbers_count = df['isValidMobile'].sum()

print(valid_phone_numbers_count)
```

```python

# Normalizing the data
df = pd.json_normalize(data, record_path=['consultationData', 'medicines'],
                       meta=['_id', 'appointmentId',
                             ['patientDetails', '_id'],
                             ['patientDetails', 'firstName'],
                             ['patientDetails', 'lastName'],
                             ['patientDetails', 'emailId'],
                             ['patientDetails', 'gender'],
                             ['patientDetails', 'birthDate'],
                             ['patientDetails', 'phrId'],
                             'phoneNumber'],
                       errors='ignore')

# Renaming columns for easier access
df.columns = ['medicineId', 'medicineName', 'medicineNameStrengthType', 'frequency',
              'duration', 'durationIn', 'instruction', 'isActive', '_id', 'appointmentId',
              'patientId', 'firstName', 'lastName', 'emailId', 'gender', 'birthDate', 'phrId', 'phoneNumber']

# Converting birthDate to datetime
df['birthDate'] = pd.to_datetime(df['birthDate'], errors='coerce')

# Calculating age
df['age'] = df['birthDate'].apply(lambda x: datetime.now().year - x.year if pd.notnull(x) else None)

# Group by patientId and calculating the number of prescribed medicines
medicine_counts = df.groupby('patientId').size().reset_index(name='num_medicines')

# Merging with age information
age_info = df[['patientId', 'age']].drop_duplicates()
merged_df = pd.merge(medicine_counts, age_info, on='patientId')

# Pearson correlation
correlation = merged_df['num_medicines'].corr(merged_df['age'], method='pearson')
correlation = round(correlation, 2)

print(correlation)
```
